```python
import pandas as pd

df = pd.read_excel('../Data/Fruentimmer_register_split.xlsx')
```

```python
import dspy

lm = dspy.LM("lm_studio/gemma-3-27b", api_base='http://127.0.0.1:1234/v1', api_key="NOT-NEEDED")
dspy.configure(lm=lm)

class ExtractPersonData(dspy.Signature):
    """Extract biographical information from text."""
    person_info: str | None = dspy.InputField()
    name: str | None = dspy.OutputField()
    origins: str | None = dspy.OutputField(desc="This can be birthplace or province/country of origin")
    age_or_date_of_birth: str | None = dspy.OutputField(desc="This can be both the person's age or their date of birth")

classify = dspy.Predict(ExtractPersonData)

results = []
for idx, row in df.iterrows():
    try:
        # Changed 'text' to 'person_info' to match your InputField name
        response = classify(person_info=row['Text'])
        
        # Store all extracted fields, not just 'result'
        results.append({
            'name': response.name,
            'origins': response.origins,
            'age_or_date_of_birth': response.age_or_date_of_birth
        })
        print(f"Processed {idx + 1}/{len(df)}")
    except Exception as e:
        print(f"Error on row {idx}: {e}")
        results.append({
            'name': None,
            'origins': None,
            'age_or_date_of_birth': None
        })

# Convert results to dataframe columns
results_df = pd.DataFrame(results)
df = pd.concat([df, results_df], axis=1)

df.to_csv('fruentimmer_struktureret.csv', index=False)
print("Done!")
```
